```python
# Import necessary packages
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd

```

```python
# Set up Splinter
browser = Browser('chrome')

# Visit the Quotes to Scrape site
url = """
    https://www.zillow.com/phoenix-az/open-house/?searchQueryState=%7B%22
    isMapVisible%22%3Afalse%2C%22mapBounds%22%3A%7B%22north%22%3A33.92057%2C%22south%22%3A33.29026%2C%22
    east%22%3A-111.717401%2C%22west%22%3A-112.464432%7D%2C%22filterState%22%3A%7B%22sort%22%3A%7B%22
    value%22%3A%22globalrelevanceex%22%7D%2C%22ah%22%3A%7B%22value%22%3Atrue%7D%2C%22open%22%3A%7B%22
    value%22%3Atrue%7D%7D%2C%22isListVisible%22%3Atrue%2C%22mapZoom%22%3A8%2C%22regionSelection%22%3A%5B%7B%22
    regionId%22%3A40326%2C%22regionType%22%3A6%7D%2C%7B%22regionId%22%3A47958%2C%22regionType%22%3A6%7D%2C%7B%22
    regionId%22%3A54346%2C%22regionType%22%3A6%7D%2C%7B%22regionId%22%3A18298%2C%22regionType%22%3A6%7D%2C%7B%22
    regionId%22%3A40298%2C%22regionType%22%3A6%7D%5D%2C%22pagination%22%3A%7B%7D%7D
    """
browser.visit(url)
# Optional delay for loading the page
browser.is_element_present_by_css('div.data-zrr-key', wait_time=5)
```

```python
# Parse the HTML
html = browser.html

# Create a BeautifulSoup object from the scraped HTML
soup = BeautifulSoup(html, 'html.parser')
```
